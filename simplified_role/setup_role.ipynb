```python
import pandas as pd
```

```python
df = pd.read_csv('source.csv')
df = df[pd.to_numeric(df["Person Username"], errors='coerce').notnull()]
df.head()
```

```python
df.info()
```

```python
df["Student Certification Expiration Date"] = pd.to_datetime(df["Student Certification Expiration Date"])
df["Student Certification Expiration Date"].replace(pd.NaT, "01/01/1901", inplace=True)
df.info()
```

```python
df.head()
```

```python
df.describe(include='all').T
```

```python
df2 = df[["Person Role", "Person Full Name", "Student Certification Expiration Date", "Certification Completion Percentage", "Certification Name", "Person Username"]]
df2.head()
```

```python
df3 = df2.groupby(['Person Role', 'Person Full Name', 'Person Username']).agg({'Student Certification Expiration Date':'min'})
df3["Student Certification Expiration Date"] = df3["Student Certification Expiration Date"].replace(pd.NaT, "Not Acquired")
df3["Student Certification Expiration Date"] = df3["Student Certification Expiration Date"].replace("1901-01-01", "Not Acquired")
df3.head(200)
df3.to_csv('expanded-READABLE.csv')
```

```python
df4 = df3.reset_index()
df4.to_csv('data_only.csv')
df4.head()
```
